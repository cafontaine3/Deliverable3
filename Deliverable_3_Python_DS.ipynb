```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris = iris.drop('species', axis = 1)
```

```python
# Add sum columm
iris['sepal_sum'] = iris['sepal_length'] + iris['sepal_width']
iris['petal_sum'] = iris['petal_length'] + iris['petal_width']
iris.head()
```

```python
# Add stats

mean = list(iris.mean())
std = list(iris.std())
min = list(iris.min())
max = list(iris.max())

# create dict for stats
summary_dict = {'mean': mean, 'std': std, 'min' : min, 'max' : max}

# create data frame for stats
summary_df = pd.DataFrame(data = summary_dict)

# drop duplicate row
summary_df = summary_df.drop([4])

# change row names
summary_df.index = ['speal_length', 'sepal_width', 'petal_length','petal_width', 'sepal_sum', 'petal_sum']

# show
print(summary_df)
```
